```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, coint
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from google.colab import files
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

# Funcin para subir archivos
def upload_and_read_excel():
    file_names = ['fro_1min', 'fro_5min', 'pep_1min', 'pep_5min', 'tsla_1min', 'tsla_5min', 'pfe_1min', 'pfe_5min']
    data_dict = {}
    for expected_filename in file_names:
        print(f"Please upload {expected_filename}.xlsx")
        uploaded = files.upload()
        matching_file = None
        for uploaded_file in uploaded.keys():
            simplified_uploaded = uploaded_file.lower().replace(' ', '').replace('(', '').replace(')', '').replace('-', '').replace('.xlsx','').replace('_','')
            simplified_expected = expected_filename.lower().replace('_','')
            if simplified_expected in simplified_uploaded:
                matching_file = uploaded_file
                break
        if matching_file:
            df = pd.read_excel(matching_file)
            if 'Close' not in df.columns:
                print(f"Error: 'Close' column not found in {matching_file}")
                return None
            close_series = df['Close'].dropna()
            data_dict[expected_filename] = close_series
            print(f"Successfully uploaded {matching_file} as {expected_filename}")
        else:
            print(f"Error: Expected file for {expected_filename} not found. Please try again.")
            return None
    return data_dict

# ADF Test con diferenciacin automtica
def adf_test_and_differentiate(series, name, max_diff=3):
    print(f"\nADF Test for {name}:")
    result = adfuller(series)
    print(f'ADF Statistic: {result[0]}, p-value: {result[1]}')
    if result[1] < 0.05:
        print(f"{name} is stationary.")
        return series, 0
    else:
        print(f"{name} is non-stationary. Differencing...")
        d = 0
        diff_series = series.copy()
        for _ in range(max_diff):
            diff_series = diff_series.diff().dropna()
            d += 1
            result = adfuller(diff_series)
            print(f"After {d} differences: ADF={result[0]}, p-value={result[1]}")
            if result[1] < 0.05:
                print(f"Stationary after {d} differences.")
                return diff_series, d
        print(f"Still non-stationary after {d} differences.")
        return diff_series, d

# Graficar ACF y PACF
def plot_acf_pacf(series, name, lags=36):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plot_acf(series, lags=lags, ax=plt.gca())
    plt.title(f'ACF for {name}')
    plt.subplot(1, 2, 2)
    plot_pacf(series, lags=lags, ax=plt.gca())
    plt.title(f'PACF for {name}')
    plt.tight_layout()
    plt.show()

# Modelos y pronsticos con evaluacin
def fit_and_forecast_models(series_original, series_stationary, name, d, forecast_steps=10):
    print(f"\nForecasting for {name}:")

    # AR sobre serie estacionaria
    ar_model = AutoReg(series_stationary, lags=2).fit()
    print("\nAR(1) Model Summary:")
    print(ar_model.summary())
    ar_fitted = ar_model.fittedvalues
    ar_mse = mean_squared_error(series_stationary[ar_fitted.index], ar_fitted)
    print(f"AR(1) In-sample MSE: {ar_mse:.4f}, AIC: {ar_model.aic:.4f}")
    ar_forecast = ar_model.forecast(steps=forecast_steps)
    # Convert AR forecast back to original scale if differenced
    ar_forecast_orig = ar_forecast if d == 0 else series_original.iloc[-1] + ar_forecast.cumsum()

    # ARMA sobre serie estacionaria
    arma_model = ARIMA(series_stationary, order=(2, 0, 2)).fit()
    print("\nARMA(1,1) Model Summary:")
    print(arma_model.summary())
    arma_fitted = arma_model.fittedvalues
    arma_mse = mean_squared_error(series_stationary[arma_fitted.index], arma_fitted)
    print(f"ARMA(1,1) In-sample MSE: {arma_mse:.4f}, AIC: {arma_model.aic:.4f}")
    arma_forecast = arma_model.forecast(steps=forecast_steps)
    # Convert ARMA forecast back to original scale if differenced
    arma_forecast_orig = arma_forecast if d == 0 else series_original.iloc[-1] + arma_forecast.cumsum()

    # ARIMA sobre serie original
    arima_model = ARIMA(series_original, order=(2, d, 2)).fit()
    print("\nARIMA(1,d,1) Model Summary:")
    print(arima_model.summary())
    arima_fitted = arima_model.fittedvalues
    arima_mse = mean_squared_error(series_original[arima_fitted.index], arima_fitted)
    print(f"ARIMA(1,d,1) In-sample MSE: {arima_mse:.4f}, AIC: {arima_model.aic:.4f}")
    arima_forecast = arima_model.forecast(steps=forecast_steps)

    # Grfico combinado para AR, ARMA, y ARIMA (serie original)
    plt.figure(figsize=(12, 6))
    plt.plot(series_original[-50:], label='Historical Original Series', color='black')
    plt.plot(np.arange(len(series_original), len(series_original)+forecast_steps), ar_forecast_orig, label='AR(1) Forecast')
    plt.plot(np.arange(len(series_original), len(series_original)+forecast_steps), arma_forecast_orig, label='ARMA(1,1) Forecast')
    plt.plot(np.arange(len(series_original), len(series_original)+forecast_steps), arima_forecast, label='ARIMA(1,d,1) Forecast')
    plt.title(f'{name} Forecasts on Original Series')
    plt.xlabel('Observations')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluacin cualitativa
    print(f"\nEvaluation of Forecasts for {name}:")
    print("- AR(2): Captures autoregressive patterns. Forecast smoothness depends on lag structure. Lower MSE/AIC indicates better fit.")
    print("- ARMA(2,2): Includes moving average component, potentially capturing more complex patterns. Compare MSE/AIC with AR.")
    print("- ARIMA(2,d,2): Models original series with differencing, suitable for non-stationary data. MSE/AIC reflects fit on price levels.")
    print("Recommendation: Choose model with lowest AIC for best balance of fit and complexity, or lowest MSE for accuracy.")

# Prueba de cointegracin
def cointegration_test(series1, series2, name1, name2):
    print(f"\nCointegration Test between {name1} and {name2}:")
    min_length = min(len(series1), len(series2))
    series1 = series1.iloc[:min_length].dropna()
    series2 = series2.iloc[:min_length].dropna()
    score, p_value, crit_values = coint(series1, series2)
    print(f'Cointegration t-statistic: {score}, p-value: {p_value}')
    print(f'Critical values: {crit_values}')
    if p_value < 0.05:
        print(f"Evidence of cointegration (p-value < 0.05).")
    else:
        print(f"No evidence of cointegration (p-value >= 0.05).")

# Prueba de cointegracin para mltiples series
def group_cointegration_test(series_list, names, interval):
    print(f"\nGroup Cointegration Test for {interval} stocks (PFE, PEP, TSLA, FRO):")
    min_length = min(len(s) for s in series_list)
    series_list = [s.iloc[:min_length].dropna() for s in series_list]
    # Test cointegration for the first series against the others combined
    score, p_value, crit_values = coint(series_list[0], pd.concat([s.to_frame() for s in series_list[1:]], axis=1))
    print(f'Cointegration t-statistic: {score}, p-value: {p_value}')
    print(f'Critical values: {crit_values}')
    if p_value < 0.05:
        print(f"Evidence of cointegration among {interval} stocks (p-value < 0.05).")
    else:
        print(f"No evidence of cointegration among {interval} stocks (p-value >= 0.05).")

# Ejecucin principal
def analyze_time_series():
    data_dict = upload_and_read_excel()
    if data_dict is None:
        print("Analysis aborted.")
        return

    # Anlisis individual de cada serie
    for name, series in data_dict.items():
        series = series.dropna()
        series.index = np.arange(1, len(series)+1)
        print(f"\n=== Analysis for {name} ===")
        stationary_series, d = adf_test_and_differentiate(series, name)
        plot_acf_pacf(stationary_series, f"{name} (Stationary)")
        fit_and_forecast_models(series, stationary_series, name, d)

    # Anlisis de cointegracin
    stocks = ['pfe', 'pep', 'tsla', 'fro']
    intervals = ['1min', '5min']
    for interval in intervals:
        print(f"\n=== Cointegration Tests for {interval} ===")
        series_list = [data_dict[f'{stock}_{interval}'].dropna().reset_index(drop=True) for stock in stocks]
        min_length = min(len(s) for s in series_list)
        series_list = [s.iloc[:min_length] for s in series_list]
        for i in range(len(stocks)):
            for j in range(i+1, len(stocks)):
                cointegration_test(series_list[i], series_list[j], f'{stocks[i]}_{interval}', f'{stocks[j]}_{interval}')
        group_cointegration_test(series_list, stocks, interval)

# Ejecutar anlisis
analyze_time_series()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
from scipy.optimize import minimize
import warnings
warnings.filterwarnings('ignore')

# Funcin para subir archivos
def upload_and_read_excel():
    file_names = ['fro_1min', 'fro_5min', 'pep_1min', 'pep_5min', 'tsla_1min', 'tsla_5min', 'pfe_1min', 'pfe_5min']
    data_dict = {}
    for expected_filename in file_names:
        print(f"Please upload {expected_filename}.xlsx")
        uploaded = files.upload()
        matching_file = None
        for uploaded_file in uploaded.keys():
            simplified_uploaded = uploaded_file.lower().replace(' ', '').replace('(', '').replace(')', '').replace('-', '').replace('.xlsx','').replace('_','')
            simplified_expected = expected_filename.lower().replace('_','')
            if simplified_expected in simplified_uploaded:
                matching_file = uploaded_file
                break
        if matching_file:
            df = pd.read_excel(matching_file)
            if 'Close' not in df.columns:
                print(f"Error: 'Close' column not found in {matching_file}")
                return None
            close_series = df['Close'].dropna()
            data_dict[expected_filename] = close_series
            print(f"Successfully uploaded {matching_file} as {expected_filename}")
        else:
            print(f"Error: Expected file for {expected_filename} not found. Please try again.")
            return None
    return data_dict

# Funcin para evaluar cada accin con criterios ampliados
def evaluate_stock(series, name, pe_ratio):
    returns = np.log(series / series.shift(1)).dropna()
    volatility = returns.std() * np.sqrt(252 * 390)

    X = np.arange(len(series)).reshape(-1, 1)
    y = series.values
    reg = LinearRegression().fit(X, y)
    slope = reg.coef_[0]

    decision = "Hold"
    reason = []

    # Evaluacin cuantitativa
    if volatility > 0.5:
        reason.append(f"Alta volatilidad ({volatility:.4f}) sugiere riesgo.")
    else:
        reason.append(f"Volatilidad moderada ({volatility:.4f}) es aceptable.")

    if slope > 0:
        reason.append(f"Tendencia positiva (pendiente={slope:.4f}) favorece compra.")
    else:
        reason.append(f"Tendencia negativa (pendiente={slope:.4f}) favorece venta en corto.")

    if pe_ratio < 15:
        reason.append(f"Ratio P/E bajo ({pe_ratio}) indica posible infravaloracin.")
    elif pe_ratio > 30:
        reason.append(f"Ratio P/E alto ({pe_ratio}) indica posible sobrevaloracin.")
    else:
        reason.append(f"Ratio P/E moderado ({pe_ratio}) es neutral.")

    # Evaluacin cualitativa y sectorial
    if 'fro' in name:
        reason.append("FRO est en el sector naviero, beneficiado por la recuperacin del comercio global.")
        reason.append("Los bajos mltiplos de valoracin (P/E) pueden reflejar oportunidad de compra.")

    if 'tsla' in name:
        reason.append("TSLA lidera el mercado de vehculos elctricos con fuerte innovacin tecnolgica.")
        reason.append("El mercado le asigna una prima por su potencial de disrupcin a largo plazo.")

    # Reglas de decisin
    if slope > 0 and pe_ratio < 15 and volatility < 0.5:
        decision = "Buy"
    elif slope > 0 and 'tsla' in name:
        decision = "Buy"
    elif slope < 0 or pe_ratio > 30:
        decision = "Short"

    return {
        'name': name,
        'volatility': volatility,
        'slope': slope,
        'pe_ratio': pe_ratio,
        'decision': decision,
        'reason': ' '.join(reason),
        'returns': returns
    }

# Funcin para optimizar portafolio
def optimize_portfolio(returns_dict, rf_rate=0.03):
    returns_df = pd.DataFrame({key: returns_dict[key]['returns'] for key in returns_dict})
    returns_df = returns_df.dropna()

    cov_matrix = returns_df.cov() * 252 * 390
    expected_returns = returns_df.mean() * 252 * 390

    def neg_sharpe_ratio(weights):
        port_return = np.sum(expected_returns * weights)
        port_vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
        return -(port_return - rf_rate) / port_vol

    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = [(0, 1) for _ in range(len(returns_dict))]
    init_guess = np.array([1./len(returns_dict)] * len(returns_dict))

    opt_result = minimize(neg_sharpe_ratio, init_guess, method='SLSQP', bounds=bounds, constraints=constraints)
    opt_weights = opt_result.x

    port_return = np.sum(expected_returns * opt_weights)
    port_vol = np.sqrt(np.dot(opt_weights.T, np.dot(cov_matrix, opt_weights)))
    sharpe_ratio = (port_return - rf_rate) / port_vol

    return {
        'weights': dict(zip(returns_dict.keys(), opt_weights)),
        'return': port_return,
        'risk': port_vol,
        'sharpe_ratio': sharpe_ratio
    }

# Funcin principal
def analyze_portfolio():
    data_dict = upload_and_read_excel()
    if data_dict is None:
        print("Analysis aborted.")
        return

    pe_ratios = {
        'pfe_1min': 12.5, 'pfe_5min': 12.5,
        'pep_1min': 25.0, 'pep_5min': 25.0,
        'tsla_1min': 80.0, 'tsla_5min': 80.0,
        'fro_1min': 8.0, 'fro_5min': 8.0
    }

    evaluation_results = {}
    for name, series in data_dict.items():
        series.index = np.arange(1, len(series)+1)
        result = evaluate_stock(series, name, pe_ratios[name])
        evaluation_results[name] = result

    print("\n=== Asset Selection Decisions ===")
    for result in evaluation_results.values():
        print(f"\n Stock: {result['name']}")
        print(f"Volatility: {result['volatility']:.4f}")
        print(f"Regression Slope: {result['slope']:.4f}")
        print(f"P/E Ratio: {result['pe_ratio']}")
        print(f" Decision: {result['decision']}")
        print(f" Reasoning: {result['reason']}")

    portfolio = optimize_portfolio(evaluation_results, rf_rate=0.03)

    print("\n=== Optimized Portfolio ===")
    print("Weights:")
    for stock, weight in portfolio['weights'].items():
        print(f"{stock}: {weight:.4f}")
    print(f"\nExpected Annual Return: {portfolio['return']:.4f}")
    print(f"Annual Risk (Volatility): {portfolio['risk']:.4f}")
    print(f"Sharpe Ratio: {portfolio['sharpe_ratio']:.4f}")

    plt.figure(figsize=(8, 6))
    plt.bar(portfolio['weights'].keys(), portfolio['weights'].values())
    plt.title('Portfolio Weights')
    plt.xlabel('Stocks')
    plt.ylabel('Weight')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Ejecutar anlisis
analyze_portfolio()
```
